```python
#importing necessary data frames
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('Video Game Reviews.csv')
```

```python
#cleaning the data
df.dropna()
df.loc[ df.duplicated() ].sort_values(by='Game Title')
df.drop_duplicates(inplace=True)
```

```python
#cleaing the data pt 2
df['Game Title'].unique()
df['User Rating'].unique()
df['Age Group Targeted'].unique()
df['Game Length (Hours)'].unique()
df['Price'].unique()
```

```python
#sorting the data
df_sorted = df.sort_values(by='Price', ascending=False)
#grouping the data
df_grouped = df.groupby('Game Title')['User Rating'].mean()
df_grouped2 = df.groupby('Game Length (Hours)')['Price'].mean()
df_grouped2 = df_grouped2.reset_index()
```

```python
#Visualizing the Data with aggregated data
sns.scatterplot(x='Price', y='Game Length (Hours)', data=df_grouped2)
plt.show()
```

```python
sns.histplot(x='Min Number of Players', data=df)
plt.show()
```

```python
#descriptive statistics
df_sorted.describe()
```

```python
#the purpose of fidning a correlation between Price and User Rating is theoretically, if the user raiting is good then the price should be higher.
#In other words, if a product is of good quality, then the price should be a little more expensive to relfect that quality.
df_sorted['Price'].corr(df_sorted['User Rating'])
```

```python
#When creating games, the price is determined by how much effort is put into the project(how many animators, how many designers etc..) therefore, a longer game should have more effort and a higher price tag.
df_sorted['Price'].corr(df_sorted['Game Length (Hours)'])
```
